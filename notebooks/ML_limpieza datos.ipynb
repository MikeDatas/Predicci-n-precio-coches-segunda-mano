```python
# Después de realizar diferentes pruebas decido que para eliminar los duplicado lo mejor es no especificar muchas columnas, ya que puede haber errores tipográficos al subir los anuncios y eso haría que no los contásemos como duplicados.
df_sin_duplicados = df.drop_duplicates(subset=['make', 'model','description', 'price'])
df_sin_duplicados
```

```python
# Eliminamos las filas con missing values en km

df_clean = df_sin_duplicados.dropna(subset=['kms'])
```

```python
# Eliminamos las filas con missing values en power

df_clean = df_clean.dropna(subset=['power'])
```

```python
# Los missing values en la columna de fuel ponemos unknown

df_clean['fuel'] = df_clean['fuel'].fillna('Otros')
```

```python
# Los missing values en la columna de shift ponemos unknown

df_clean['shift'] = df_clean['shift'].fillna('Unknown')
```

```python
# Eliminamos columan que no aporten valor al estudio

df_clean = df_clean.drop(columns=["vehicle_type","version","doors","color","photos","description","currency","location","publish_date","update_date","dealer_name","dealer_description","dealer_address","dealer_city","dealer_country_code","dealer_is_professional","dealer_website","dealer_registered_at","date","dealer_zip_code"])
```

```python
# Se realizan Boxplot

# Lista de columnas numéricas 
columnas_numericas = ['year', 'kms', 'power', 'price']

# Configurar tamaño: 1 gráfico por fila
plt.figure(figsize=(12, 6 * len(columnas_numericas)))

for i, col in enumerate(columnas_numericas, 1):
    plt.subplot(len(columnas_numericas), 1, i)
    sns.boxplot(data=df_clean, x=col, color='skyblue')
    plt.title(f'Boxplot para {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```
