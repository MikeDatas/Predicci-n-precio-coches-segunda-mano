```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data_limpio.csv")
```

```python
# Crear los countplots para columnas categóricas
columnas_categoricas = ['make', "model", 'fuel', 'shift']

# Configurar la figura de matplotlib (más ancha y con una fila por gráfico)
plt.figure(figsize=(12, 6 * len(columnas_categoricas)))  # Ajusta el tamaño en función del número de columnas

# Crear countplots para cada columna categórica
for i, col in enumerate(columnas_categoricas, 1):
    plt.subplot(len(columnas_categoricas), 1, i)  # Una columna por gráfico
    sns.countplot(data=df, x=col, palette="viridis")
    plt.title(f'Countplot para {col}')
    plt.xticks(rotation=90)  # Rotar las etiquetas del eje x si es necesario

plt.tight_layout()  # Ajustar el layout para evitar solapamientos
plt.show()
```

```python
# Crear histogramas para columnas numéricas

columnas_numericas = ['year', 'kms', 'power', 'price']

# Configurar la figura de matplotlib: 1 gráfico por fila, más alto
plt.figure(figsize=(12, 6 * len(columnas_numericas)))  # ancho 12, alto dinámico

# Crear histogramas para cada columna numérica
for i, col in enumerate(columnas_numericas, 1):
    plt.subplot(len(columnas_numericas), 1, i)  # una fila por gráfica
    sns.histplot(df[col], kde=True, color="skyblue")
    plt.title(f'Histograma para {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')

plt.tight_layout()  # Ajustar el diseño
plt.show()
```

```python
# Lista de columnas numéricas 
columnas_numericas = ['year', 'kms', 'power', 'price']

# Configurar tamaño: 1 gráfico por fila
plt.figure(figsize=(12, 6 * len(columnas_numericas)))

for i, col in enumerate(columnas_numericas, 1):
    plt.subplot(len(columnas_numericas), 1, i)
    sns.boxplot(data=df, x=col, color='skyblue')
    plt.title(f'Boxplot para {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)

# Imprimir la matriz de correlación
print(corr_matrix)

# Visualizar la matriz de correlación usando un heatmap
plt.figure(figsize=(10, 8))  # Tamaño de la figura
sns.heatmap(corr_matrix, annot=True, vmin = -1,cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()

```

```python
# Asegúrate de que el DataFrame df contiene esas columnas
columnas_numericas = ['year', 'kms', 'power', 'price']
df_pairplot = df[columnas_numericas].copy()

# Crear pairplot
sns.pairplot(df_pairplot, diag_kind='kde')  # diag_kind='hist' también es válido
plt.suptitle("Pairplot de variables numéricas", y=1.02)
plt.show()
```
